### Observation

The dataset contains 540 patient records with clinical, hormonal, and lifestyle-related features associated with PCOS prediction.

Key observations from the statistical summary:

- The average age of patients is around 31 years.
- Average BMI is approximately 24, indicating most patients fall near the normal to overweight range.
- Weight ranges from 31 kg to 108 kg, showing significant variation among individuals.
- Height ranges between 137 cm and 180 cm.
- PCOS values are binary (0 = No PCOS, 1 = PCOS), confirming this is a classification problem.
- Several healthcare measurements such as hormone levels, blood pressure, and follicle counts show wide variability, which may influence prediction performance.

The statistical summary helps understand feature distribution, detect unusual values, and prepare the dataset for exploratory analysis and machine learning.
